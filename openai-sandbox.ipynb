# Simple agent

````python
import gym

def run_episode():
    env = gym.make('CartPole-v0')
    state = env.reset()
    steps = 0
    
    while True:
        env.render()
        
        [x, vx, a, va] = state
        action = 0 if a < 0 else 1
        
        (state, reward, done, _) = env.step(action)
        steps += 1
        if done:
            break
            
    env.close()
    print('Finished in %s steps' % steps)
````

````python
run_episode()
````
